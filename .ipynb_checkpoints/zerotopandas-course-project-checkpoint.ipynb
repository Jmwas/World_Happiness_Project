# World Happiness Report

 
In this project, I will analyze the data and figure out what brings about world happiness and other related factors.

I will use various python libraries to perform the analysis and visualize the output including pandas, numpy, seaborn and matplotlib
    
The World Happiness Report is a publication of the United Nations Sustainable Development Solutions Network. It contains rankings of national happiness, based on Gallup poll respondent ratings of their own lives. Respondents of different countries are asked to think of a ladder, with the best possible life for them being a 10, and the worst possible life being a 0. They are then asked to rate their own current lives on that 0 to 10 scale. The report correlates the life evaluation results with various life factors.

From the report, the factors that are used to determine happiness are:

>- GDP per capita
>- Social support
>- Healthy life expectancy at birth
>- Freedom to make life choices
>- Generosity
>- Perceptions of corruption


The data was acquired from Kaggle

- Determine if there was a shift in life ladder for countries between the years 2015 - 2020
- I will split the countries in different groups 

> - Low - countries with a life ladder in the 20% or lower
> - Lower than average - countries with a life ladder in the range between 21% - 40% 
> - Average - countries with a life ladder in the range between 41% - 60% 
> - Higher than average - countries with a life ladder in the range between 61% - 75% 
> - High - Countries with a life ladder higher than 75%

```python
# splitting data into groups
data_2015 = wh_df2[wh_df2['year'] == 2015]['Life Ladder']

lo_2015 = np.round(np.percentile(data_2015, 20), 2) # 20th percentile
md_lo_2015 = np.round(np.percentile(data_2015, 40), 2) # 40th percentile
md_2015 = np.round(np.percentile(data_2015, 60), 2) # 60th percentlie
md_hi_2015 = np.round(np.percentile(data_2015, 75), 2) #75th percentile
hi_2015 = np.round(np.percentile(data_2015, 76), 2) # 76th percentile
```

```python
low_2015, mid_low_2015, mid_2015, mid_high_2015, high_2015  = 0,0,0,0,0

for i in data_2015:
    if i <= lo_2015:
        low_2015 += 1
    elif i <= md_lo_2015:
        mid_low_2015 += 1        
    elif i <= md_2015:
        mid_2015 += 1
    elif i <= md_hi_2015:
        mid_high_2015 += 1
    else:
        high_2015 += 1
```

```python
data_2018 = wh_df2[wh_df2['year'] == 2018]['Life Ladder']

lo_2018 = np.round(np.percentile(data_2018, 20) , 2)
md_lo_2018 = np.round(np.percentile(data_2018, 40), 2)
md_2018 = np.round(np.percentile(data_2018, 60), 2)
md_hi_2018 = np.round(np.percentile(data_2018, 75), 2)
hi_2018 = np.round(np.percentile(data_2018, 76), 2)
```

#### Q1: Are people higher on the life ladder more generous?
> Generosity is described as the residual of regressing the national average of responses to the question "Have you donated money to a charity in the past month?"

```python
x = 'Life Ladder'
y = 'Generosity'
sns.set(rc={"figure.figsize":(16, 8)})
plt.title('Generosity vs Life Ladder?')

# plot specifying the x, y parameters to avoid getting misinterpretation warning 
sns.scatterplot(x=x, y=y, data=wh_df2, s=90, alpha=0.9, hue='year', legend="full"); 
```

From the scatterplot, we can  make a few observations
> - For respondents with a life ladder below 6, there is almost an even split between people who had donated to charity to those who hadn't
> - For those above 6, though less in count than those below, there are slightly more respondents who gave to charity than those who did not donate.
> - There is a few outliers as we can see the most charitable (above 0.6 in generosity) were between 4 and 5 in the life ladder scale.

We can therefore say that respondents on a higher life ladder are not more generous than those who have a lower life ladder.

#### Q2: What attributes mostly affects happiness?
- To figure this out, we can use a correlation heatmap.
- Correlation is a term used to represent the statistical measure of linear relationship between two variables. It can also be defined as the measure of dependence between two different variables. 

```python
# Select only the columns that we need for the heatmap
wh_df3 = wh_df2.loc[:,'Life Ladder':]
```

```python
plt.figure(figsize=(12,8))
plt.title("Life Happiness Heatmap", fontsize =20)

# np.triu() isolates the upper triangle of a matrix while turning all values in the lower triangle to 0
# np.ones_like changes all the isolated values to 1
# mask hides a cell where the value of the cell is 1 
mask = np.triu(np.ones_like(wh_df3.corr(), dtype = bool))

heatmap = sns.heatmap(wh_df3.corr(), annot=True, cmap='coolwarm', mask = mask);
```

> ###### On the legend, a red color indicates positive correlation which means if one attribute goes up and the other variable also goes up and vice versa. 
> ###### Blue indicates a negative correlation which means that if one attribute decreases in value, the other increases in value or vice versa

Looking at the heatmap, we can make a few observations:
- The attribute that has the highest positive correlation is GDP per capita followed closely by health life expectancy at birth. The higher the GDP per capita and health life expectancy, the more happier the respondents are.
- Perceptions of corruption have the highest negative impact on life ladder. The more the perception of corruption, the lower the life ladder.
- We can see that GDP per capita and health life expectancy have the most positive correlation while perceptions of corruption and freedom to make life choices have the most negative correlation. 


###### Q3: What is the effect of per capita GDP on corruption?
> To figure this out, we can use the same correlation heatmap that we used above.

```python
plt.title("Effect of GPD on Corruption", fontsize =20)
heatmap = sns.heatmap(wh_df3.corr(), annot=True, cmap='coolwarm', mask = mask);
```

- If we look at the correlation between GDP and perceptions corruption, we see that there is a negative correlation.
- That means, higher levels of corruption, or the more the respondents think their country is corrupt, the less the GDP per capita and vice versa.

#### Q4: Are Americans free to make their own life choices?

> Freedom to make life choices is described as the national average of responses to the question "Are you satisfied or dissatisfied with your freedom to choose what you do with your life?"

Based on this description, let's see if Americans think they are free to make decisions and how the response has changed over the years and compare that to the world's average and to Finland which topped the list in the life ladder in 2020

```python
# Selecting data only for the USA
freedom_usa_df = wh_df2[wh_df2['Country name']=='United States']
freedom_usa_df.head()
```

```python
# Selecting data only for Finland where year is after 2007
freedom_fin_df = wh_df2[(wh_df2['Country name']=='Finland') & (wh_df2['year']>2007)]
freedom_fin_df.head()
```

```python
# create dataframe of 
freedom_world_df = wh_df2[wh_df2['year']>2007].groupby('year').mean() 
# freedom_world_df = wh_df2[wh_df2['year']>=2008].groupby('year').mean() # alternate code
freedom_world_df.head()
```

- Looking at the dataframe, the year starts in 2008. I will have to match the starting years for the rest of the world dataframe otherwise there will be an error when plotting the two plots. 
- To do that, I will select data with the year greater than 2007.

- Looking at the three dataframes, we can see that the year and the freedom to make life choices columns are not on the same index and hence will throw a ValueError when we plot them. 
- To fix this, I will use reset_index() and set_index()

```python
freedom_usa_df = freedom_usa_df.loc[:,'year':] # slicing dataframe to remove country name
freedom_usa_df.reset_index(drop=['index'], inplace=True) # reset the index column and remove it from the dataframe
freedom_usa_df.head()
```

```python
freedom_fin_df = freedom_fin_df.loc[:,'year':] # slicing dataframe to remove country name
freedom_fin_df.reset_index(drop=['index'], inplace=True) # reset the index column and remove it from the dataframe
freedom_fin_df.head()
```

```python
freedom_world_df.reset_index(inplace=True)
freedom_world_df.head()
```

```python
# setting the width and height of the plot
f = plt.figure()
f.set_figwidth(12)
f.set_figheight(5)

# naming the x and y axis
plt.xlabel('year')
plt.ylabel('Freedom To Make Life Choices')

plt.plot('year', 'Freedom to make life choices', data=freedom_usa_df, label='USA')
plt.plot('year', 'Freedom to make life choices', data=freedom_fin_df, label='Finland')
plt.plot('year', 'Freedom to make life choices', data=freedom_world_df, label='Worldwide Average)')

plt.legend()
plt.title('Freedom to Make Life Choices (USA vs. Rest of the World) ');
```

From the results, we can make a few observations:
- Compared to the world's average, America has more freedom to make life choices but it's not as high as the Finns. 
- The level of satisfaction has been on a steady decrease as compared to Finland and the world's average which have been on a steady increase.
- There is a noticeable sharp increase or decrease in the levels of freedom to make life choices which is interesting but unfortunately the data doesn't give reasons for such.

#### Q5: On average, which country has been the happiest between 2015 and 2020?

To figure this, I will select only the countries with data between 2015 and 2020 and find the average life ladder for these countries. 

```python
# select data for years between 2015 and 2020
happy_df = wh_df2[wh_df2['year'].between(2015, 2020)]

# find the average for the life ladder in each individual country
happy_ll_df = happy_df.groupby('Country name', as_index=False)['Life Ladder'].mean().round(3)
happy_ll_df.sort_values(by='Life Ladder',ascending=False).head(1)
```

```python
data = happy_ll_df.sort_values(by='Life Ladder',ascending=False)

sns.set(rc={"figure.figsize":(15, 30)}) # set the size of the grid 
sns.set_palette("pastel")
sns.barplot(x='Life Ladder', y='Country name', data=data)
plt.title('Five year Life Ladder average ');
```

On average, we can see that Finland has the highest life ladder for the past five years and on the other end, Yemen has the lowest average Life Ladder for the past five years.
